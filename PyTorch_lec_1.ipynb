<h3 style="color:#2a7ae2;">🚀 PyTorch Device Setup </h3>

-   **`torch.cuda.is_available()`**:
    -   Checks for the presence and accessibility of a CUDA-enabled GPU.
    -   Returns `True` if a compatible NVIDIA GPU and drivers/CUDA toolkit are installed.

-   **`'cuda' if torch.cuda.is_available() else 'cpu'`**:
    -   A Python conditional expression.
    -   Evaluates to the string `'cuda'` if a GPU is available.
    -   Evaluates to the string `'cpu'` if no GPU is available.

-   **`torch.device(...)`**:
    -   Constructs a `torch.device` object based on the conditional expression.
    -   This `device` variable then specifies where PyTorch tensors and models should reside (GPU or CPU).

This setup allows your code to:
-   Leverage GPU acceleration when available for significant speedups.
-   Automatically fall back to CPU when a GPU is not found, ensuring portability.

```python
#Creating a new tensor on GPU

torch.rand((2,3),device=device)
```

```python
# Moving an Existing Tensor to GPU
a=torch.rand((2,3))# created on CPU
b=a.to(device) # Moved to Gpu on a different name
print(a,'\n',b)
```

```python
import time

# Define the size of the matrices
size = 10000 # Large size for performance comparison

# Create random matrices on CPU
matrix_cpu1 = torch.randn(size, size)
matrix_cpu2 = torch.randn(size, size)

# Measure time on CPU
start_time = time.time()
result_cpu = torch.matmul(matrix_cpu1, matrix_cpu2)# Matrix Mul on CPU
cpu_time = time.time() - start_time
print(f"Time on CPU: {cpu_time:.4f} seconds")

# Move matrices to GPU
matrix_gpu1 = matrix_cpu1.to('cuda')
matrix_gpu2 = matrix_cpu2.to('cuda')

# Measure time on GPU
start_time = time.time()
result_gpu = torch.matmul(matrix_gpu1, matrix_gpu2) # Matrix multiplication on GPU
torch.cuda.synchronize() # Ensure all GPU operations are complete
gpu_time = time.time() - start_time
print(f"Time on GPU: {gpu_time:.4f} seconds")

#Compare the Result
print("\nSpeedup:", cpu_time / gpu_time)
```

### 🔄 Tensor Manipulation: `Permute`, `Squeeze`, and `Unsqueeze`

These functions are essential for changing the shape and arrangement of tensors without altering their underlying data. They are crucial for preparing data for models, matching expected input shapes, and handling batch dimensions.

---

#### 1. `torch.permute()`

*   **What it does:** Rearranges the dimensions (axes) of a tensor according to a specified order. It's like transposing a 2D matrix, but for any number of dimensions.
*   **How it works:** You provide a tuple or list of integers representing the new order of dimensions. For example, if a tensor has shape `(D0, D1, D2)`, `tensor.permute(1, 2, 0)` would change its shape to `(D1, D2, D0)`.
*   **When to use it:**
    *   **Changing Image Data Layout:** Many image processing tasks and models (especially in PyTorch) expect image data in `(Batch, Channels, Height, Width)` format, while common loading libraries might produce `(Batch, Height, Width, Channels)`. `permute` is used to switch `HWC` to `CHW` (e.g., `image.permute(0, 3, 1, 2)`).
    *   **Aligning Dimensions:** When performing operations between tensors that require specific dimension alignment (e.g., matrix multiplication, broadcasting).

---

#### 2. `torch.squeeze()`

*   **What it does:** Removes all dimensions of size 1 from the tensor's shape.
*   **How it works:** If a tensor has a shape like `(1, 3, 1, 256, 1)`, `squeeze()` would result in a shape of `(3, 256)`. You can also specify a `dim` argument to remove a dimension of size 1 at a particular index (e.g., `tensor.squeeze(0)` would remove the first dimension if it's size 1).
*   **When to use it:**
    *   **Removing Redundant Batch Dimensions:** Often, when processing a single item, you might add a batch dimension of size 1 (e.g., `(1, Channels, Height, Width)`). After processing, you might want to remove this singleton dimension to get `(Channels, Height, Width)`.
    *   **Cleaning Up Shapes:** Simplifying tensor shapes after operations that might introduce unnecessary singleton dimensions.

---

#### 3. `torch.unsqueeze()`

*   **What it does:** Adds a dimension of size 1 to the tensor's shape at a specified position.
*   **How it works:** You provide the `dim` argument (an integer) where the new dimension should be inserted. For example, if a tensor has shape `(3, 256)`, `tensor.unsqueeze(0)` would result in `(1, 3, 256)`. `tensor.unsqueeze(1)` would result in `(3, 1, 256)`.
*   **When to use it:**
    *   **Adding Batch Dimensions:** Most neural networks expect inputs to have a batch dimension (even if it's a batch of one). If you have a single image with shape `(Channels, Height, Width)`, you'd use `image.unsqueeze(0)` to make it `(1, Channels, Height, Width)`.
    *   **Broadcasting:** Preparing tensors for broadcasting operations where one tensor needs an extra dimension to match the other's shape.
    *   **Feature Expansion:** Adding a new dimension to represent a single feature map or channel where the model expects an explicit dimension.
